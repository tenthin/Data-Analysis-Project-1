```python
import pandas as pd
```

```python
df1=pd.read_csv('raw_data.csv')
```

```python
df1.head()
```

# Checking for Null Values

```python
df1.isnull().count()
```

# Dropping the unnamed columns

```python
df1.drop(['Unnamed: 9', 'Unnamed: 10','Unnamed: 11','Unnamed: 12'], axis = 1,inplace=True) 
```

```python
df1
```

```python
df1.drop(['gdp_per_capita'], axis = 1,inplace=True) 
```

```python
df1
```

# Renaming column 'Unnamed:13' to 'gdp_per_capita'

```python
df1.rename(columns = {'Unnamed: 13':'gdp_per_capita'}, inplace = True)
```

```python
df1
```

# Mean of total cases by location

```python
df1.groupby('location').total_cases.mean()
```

# Mean of total deaths by location

```python
df1.groupby('location').total_deaths.mean()
```

# Showing only for Afghanistan

```python
df1[df1.location=='Afghanistan']
```

### I have done for Afghanistan, we can do this for different locations ( we can also use different aggregation functions like max,min etc.)

```python
df1[df1.location == 'Afghanistan'].total_cases.mean()
```

# Using multiple aggregation function at once

```python
df1.groupby('location').total_deaths.agg(['count','max','min','mean'])
```

```python
df1.groupby('location').mean()
```
